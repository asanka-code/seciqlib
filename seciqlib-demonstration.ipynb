## Seciqlib - Demonstration

This notebook demonstrates the basic use of the seciqlib library to deal with I-Q data files generated by GNURadio.

First of all, we import the library.

```python
import seciqlib
```

### Get time duration of the I-Q data file

```python
duration = seciqlib.getTimeDuration("3des.dat")
print(str(duration) + " seconds")
```

### Plotting the seleted I-Q data segment in different ways

```python
# user defined variables
cFileName = "3des.dat"
offset_time = 0
window_time = 0.01

segment = seciqlib.getSegmentData(cFileName, offset_time, window_time)
seciqlib.plotWaveform(segment)
```

```python
# user defined variables
cFileName = "3des.dat"
offset_time = 0
window_time = 0.01

segment = seciqlib.getSegmentData(cFileName, offset_time, window_time)
seciqlib.plotScatter(segment)
```

```python
# user defined variables
cFileName = "3des.dat"
offset_time = 0
window_time = 0.01

segment = seciqlib.getSegmentData(cFileName, offset_time, window_time)
seciqlib.plotPSD(segment)
```
